```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##### Standardization

```python
# Dataset: Social_Network_Ads.csv
# Location: https://mitu.co.in/dataset
```

```python
df = pd.read_csv('/home/mitu/Social_Network_Ads.csv')
```

```python
df
```

```python
# store 'Age' and 'EstimatedSalary' in x variable.
x = df[['Age','EstimatedSalary']]
```

```python
x.describe()
```

```python
plt.scatter(df['Age'], df['EstimatedSalary'], c = df['Purchased'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_std = scaler.fit_transform(x)
```

```python
x_std
```

```python
x_std = pd.DataFrame(x_std, columns=x.columns)
```

```python
x_std
```

```python
x_std.describe()
```

```python
plt.subplot(1,2,1)
plt.title('Original Data')
plt.scatter(df['Age'], df['EstimatedSalary'], c = df['Purchased'])

plt.subplot(1,2,2)
plt.title('Standardized Data')
plt.scatter(x_std['Age'], x_std['EstimatedSalary'], c = df['Purchased'])
```

##### Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_norm = scaler.fit_transform(x)
```

```python
x_norm
```

```python
x_norm = pd.DataFrame(x_norm, columns=x.columns)
```

```python
x_norm.describe()
```

```python
plt.subplot(1,2,1)
plt.title('Original Data')
plt.scatter(df['Age'], df['EstimatedSalary'], c = df['Purchased'])

plt.subplot(1,2,2)
plt.title('Normalized Data')
plt.scatter(x_norm['Age'], x_norm['EstimatedSalary'], c = df['Purchased'])
```
