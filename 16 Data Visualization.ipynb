##### Bar plot

```python
enrol = pd.DataFrame({
    'Students': [67,89,106,179,47,122,100],
    'Languages': ['C','C++','Java','Python','R','PHP','JavaScript']
})
```

```python
enrol
```

```python
plt.title('Student Enrollment for Training')
plt.xlabel('Language')
plt.ylabel('Total Students')
plt.bar(enrol['Languages'], enrol['Students'], color = 'magenta')
```

```python
plt.title('Student Enrollment for Training')
plt.xlabel('Language')
plt.ylabel('Total Students')
plt.bar(enrol['Languages'], enrol['Students'], color = 'cyan', )
plt.plot(enrol['Languages'], enrol['Students'], marker = '*')
```

```python
plt.title('Student Enrollment for Training')
plt.ylabel('Language')
plt.xlabel('Total Students')
plt.barh(enrol['Languages'], enrol['Students'], color = 'purple', )
```

```python
enrol = pd.DataFrame({
    'Students': [67,89,106,179,47,122,100],
    'Languages': ['C','C++','Java','Python','R','PHP','JavaScript'],
    'AvgMarks': [67.45, 69.23, 72.45, 78.23, 63.27, 69.34, 74.23]
})
```

```python
enrol
```

```python
plt.title('Student Enrollment for Training')
plt.xlabel('Language')
plt.ylabel('Total Students')
plt.bar(enrol['Languages'], enrol['Students'], color = 'teal', 
        align='edge', width=0.3, label='Total Student')
plt.bar(enrol['Languages'], enrol['AvgMarks'], color = 'lightgreen', 
        align='center', width=0.2, label='AvgMarks')
plt.legend()
```

##### Pie Chart

```python
plt.figure(figsize=(6,6))
plt.title('% Share of students')
plt.pie(enrol['Students'], labels=enrol['Languages'], autopct='%2.2f%%');
```
