##### Import the libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

##### Load the data

```python
df = pd.read_csv('petrol_consumption.csv')
```

```python
df
```

```python
df.corrwith(df['Petrol_Consumption'])
```

##### Seperate the input and output variables

```python
# input
x = df.drop('Petrol_Consumption', axis = 1)

# output
y = df['Petrol_Consumption']
```

```python
sns.kdeplot(y);
```

##### Cross - Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train
```

```python
x_test
```

##### Build the model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg = DecisionTreeRegressor(random_state=0)
```

```python
reg.fit(x_train, y_train)
```

##### Plot the tree

```python
from sklearn.tree import plot_tree
```

```python
x.columns
```

```python
plt.figure(figsize=(30,14))
plot_tree(reg, fontsize=7, filled=True, rounded=True, 
          feature_names=['tax','inc','hways','popl']);
plt.savefig('mytree.png')
```

##### Performance Evaluation

```python
y_pred = reg.predict(x_test)
```

```python
result = pd.DataFrame({
    'actual': y_test,
    'predicted': y_pred
})
```

```python
result
```

```python
# mean absolute error
np.mean(abs(y_test - y_pred))
```

```python
new = pd.DataFrame(data=[[9.00, 4258, 4686, 0.517],
                        [7.5, 5126, 14186, 0.525]],
                  columns=x.columns)
```
