```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/mitu/mtcars.csv')
```

```python
df
```

###### Separate the input and output variables

```python
# input -> disp, hp, wt
x = df[['disp','hp','wt']]
# output -> mpg
y = df['mpg']
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x, y)
```

```python
reg.score(x, y)
```

```python
x.corrwith(y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
disp = 221
hp = 102
wt = 2.91
```

```python
disp*-9.37009081e-04 + hp*-3.11565508e-02 + wt*-3.80089058 + 37.10550526903182
```

```python
reg.predict([[disp, hp, wt]])
```
