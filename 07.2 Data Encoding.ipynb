#### Data Encoding: Represent the string values in numerical form

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### Label Encoding

```python
df = pd.read_csv('student.csv')
```

```python
df
```

```python
# import the class
from sklearn.preprocessing import LabelEncoder
```

```python
# create the object
le = LabelEncoder()
```

```python
# apply the encoding
df['name'] = le.fit_transform(df['name'])
```

```python
df
```

```python
new = 'anil'
```

```python
le.transform([new])
```

```python
le.inverse_transform([5])
```

```python
df.dtypes
```
